## Programa 1
Programa que simule el lanzamiento de un dado 1000 veces. Almacena los números en un arreglo y muestra el histograma con las frecuencias de las veces que salio cada cara.

### Código

```python
diceRolls = 1000
randomNumbers = []

for i in range(diceRolls):
    number = random.randint(1, 6)
    randomNumbers.append(number)
print(randomNumbers)
```

```python
# Obtener frecuencias
frequencies = {i: randomNumbers.count(i) for i in range(1, 7)}
print(f"Frecuencias: {frequencies}")
```

### Histograma

```python
sns.histplot(data=randomNumbers, color='skyblue', bins=6, kde=False)
plt.title('Frecuencia de Números Obtenidos al Lanzar un Dado 1000 Veces')
plt.xlabel('Número Obtenido')
plt.ylabel('Frecuencia')
plt.xticks(range(1, 7))
plt.show()
```

## Programa 2
Programa que simule el lanzamiento de un dado, 1000 veces de tal forma que haya una probabilidad del 50% de que salga el 6, y el resto de los números el otro 50% de manera equitativa.  Muestra en una tabla las veces que salió cada número y aparte muéstralo como porcentaje.

### Código

```python
def forceNumber(a, b, diceRolls, forcedNumber):
    randomNumbers = []
    for i in range(diceRolls):
        if i % 2 == 0:
            randomNumbers.append(forcedNumber)
        else:
            number = random.randint(a, b)
            randomNumbers.append(number)
    return randomNumbers

randomNumbers = forceNumber(1, 5, 1000, 6)
print(randomNumbers)
print(len(randomNumbers))
```

```python
# Tabla con los resultados, numero, frecuencia y frecuencia relativa
results = {number: randomNumbers.count(number) for number in range(1, 7)}
totalRolls = len(randomNumbers)
relativeFrequencies = {number: frequency / totalRolls for number, frequency in results.items()}
print("Número | Frecuencia | Frecuencia Relativa")
for number in range(1, 7):
    print(f"{number}      | {results[number]}         | {relativeFrequencies[number]:.4f}")
```
